```python
def parity(n: int) -> int:
    """generate largest number of bit length n
    
    Args:
        n: integer to get parity of
    
    Returns:
        int 0 or 1 for parity
    """
    bl = n.bit_length()
    while bl>1:
        print(n)
        if n<len(cache):
            return cache[n]
        if (bl % 2)==0:
            s=(bl//2)
            n=(n&(2**s-1)) ^ (n>>s)
        else:
            n=(n >> 1 )^(n&1)
        bl=n.bit_length()
    return n
```
